```python
import pandas as pd

data =pd.read_csv("FootwearData.csv")
data.head(5)
```

```python
# Data Averaging (old-method vs new method)
# Create a new column for the average of old-method and new-method columns
data['Flexibility / Stiffness (average) (N)'] = data[['Flexibility / Stiffness (old method) (N)', 'Flexibility / Stiffness (new method) (N)']].mean(axis=1, skipna=True)

# Repeat the same for other old-method vs new-method columns if needed
data['Toebox width - widest part (average) (mm)'] = data[['Toebox width - widest part (old method) (mm)', 'Toebox width - widest part (new method) (mm)']].mean(axis=1, skipna=True)
data['Toebox width - big toe (average) (mm)'] = data[['Toebox width - big toe (old method) (mm)', 'Toebox width - big toe (new method) (mm)']].mean(axis=1, skipna=True)
```

```python
# Data Cleanup 
# Remove the old-method and new-method columns
data.drop(columns=['Flexibility / Stiffness (old method) (N)', 'Flexibility / Stiffness (new method) (N)',
                   'Toebox width - widest part (old method) (mm)', 'Toebox width - widest part (new method) (mm)',
                   'Toebox width - big toe (old method) (mm)', 'Toebox width - big toe (new method) (mm)'], inplace=True)

data.rename(columns={
    'Midsole softness in cold (%) (%)': 'Midsole softness in cold (%)',
    'Stiffness in cold (%) (%)': 'Stiffness in cold (%)'
}, inplace=True)
```

```python
pd.set_option('display.max_columns', None)
# print the # of columns
print(len(data.columns))
print(len(data))
data.head(20)
```

```python
# Combine our pace types into a cleaner format 
# Pace
# Daily running           265
# Tempo                    31
# Tempo, Daily running     26
# Competition              26
# Daily running, Tempo     19
# Tempo, Competition       12
# Competition, Tempo        6
# Name: count, dtype: int64

# I.e. Tempo, Daily running = Daily running, Tempo, Tempo, Competition = Competition, Tempo
# Combine the pace types into a single column

# Create a mapping to standardize pace categories
pace_mapping = {
    'Tempo, Daily running': 'Tempo',
    'Competition, Tempo': 'Competition'
}

# Apply the mapping to standardize similar categories
data['Pace'] = data['Pace'].replace(pace_mapping)

# Verify the changes
print(data['Pace'].value_counts())
```

```python
# Count the occurrences of each Strike Pattern
strike_pattern_counts = data['Strike Pattern'].value_counts()

# Display the counts
print(strike_pattern_counts)
# Create a mapping to standardize strike patterns
strike_pattern_mapping = {
    'Heel strike  |': 'Heel strike'
}

# Apply the mapping to standardize strike patterns
data['Strike Pattern'] = data['Strike Pattern'].replace(strike_pattern_mapping)

# Verify the changes
print(data['Strike Pattern'].value_counts())
```

```python
# Save the cleaned data to a new CSV file
data.to_csv("FootwearData_cleaned.csv", index=False)
```
